```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the studies
studies = ['R.A’izzah(2024)', 'S.Ganesan(2024)','D.Manolescu(2024)' ,'Proposed Study']
loss = [0.96, 0.42,0.95, 0.26]

# Setting up the bar plot
x = np.arange(len(studies))  # label locations
width = 0.6  # further increase bar width to reduce space between bars

fig, ax = plt.subplots(figsize=(7, 6))

# Plotting validation accuracy
bars1 = ax.bar(x, loss, width, label='loss', color='green')

# Add labels, title, and custom x-axis tick labels
ax.set_xlabel('Studies')
ax.set_ylabel('Scores')
ax.set_title('Comparison of loss')
ax.set_xticks(x)
ax.set_xticklabels(studies)

# Adding a legend
ax.legend()

# Adding data labels on bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.3f}',  # format the label with 3 decimal places
                    xy=(bar.get_x() + bar.get_width() / 2, height),  # adjusted position for label
                    xytext=(0, 1),  # offset above the bar
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bars1)

plt.tight_layout()
plt.show()
```
